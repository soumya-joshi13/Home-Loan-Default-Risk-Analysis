```python
## MISSING VALUE PERCENTAGE
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage = missing_percentage.sort_values(ascending=False)

missing_percentage
```

```python
## Create missing value dataframe
```

```python
missing_df = pd.DataFrame({
    'Column_Name': missing_percentage.index,
    'Missing_Percentage': missing_percentage.values
})

missing_df.head(15)
```

```python
## Columns with more than 50% missing values
```

```python
high_missing_cols = missing_df[missing_df['Missing_Percentage'] > 50]

high_missing_cols
```

```python
## Drop columns with more than 50% missing values
```

```python
cols_to_drop = high_missing_cols['Column_Name'].tolist()

df.drop(columns=cols_to_drop, inplace=True)
```

```python
## Check shape after dropping columns
```

```python
df.shape
```

```python
## Separate numerical and categorical columns
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical columns:", len(numerical_cols))
print("Categorical columns:", len(categorical_cols))
```

```python
## Missing values in numerical columns
```

```python
df[numerical_cols].isnull().sum().sort_values(ascending=False).head(10)
```

```python
## Fill numerical missing values with median
```

```python
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
## Missing values in categorical columns
```

```python
df[categorical_cols].isnull().sum().sort_values(ascending=False).head(10)
```

```python
## Fill categorical missing values with mode
```

```python
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
## Final missing value check
```

```python
df.isnull().sum().sum()
```

## LOAD BUREAU DATA

```python
bureau = pd.read_csv('bureau.csv')
```

```python
bureau.head()
```

```python
bureau.shape
```

```python
bureau.info()
```

```python
bureau['SK_ID_CURR'].nunique()
```

```python
bureau_useful_cols = [
    'SK_ID_CURR',
    'CREDIT_ACTIVE',
    'AMT_CREDIT_SUM',
    'AMT_CREDIT_SUM_DEBT',
    'AMT_CREDIT_SUM_OVERDUE',
    'DAYS_CREDIT',
    'DAYS_CREDIT_ENDDATE'
]

bureau = bureau[bureau_useful_cols]
```

```python
## Convert CREDIT_ACTIVE to numerical
```

```python
bureau['CREDIT_ACTIVE'] = bureau['CREDIT_ACTIVE'].map({
    'Active': 1,
    'Closed': 0
})
```

```python
bureau.isnull().sum()
```

```python
## Fill missing numerical values with median
```

```python
for col in bureau.columns:
    if bureau[col].dtype != 'object':
        bureau[col].fillna(bureau[col].median(), inplace=True)
```

```python
bureau.isnull().sum()
```

```python
## Aggregate bureau data at customer level
```

```python
bureau_agg = bureau.groupby('SK_ID_CURR').agg({
    'CREDIT_ACTIVE': 'mean',
    'AMT_CREDIT_SUM': 'sum',
    'AMT_CREDIT_SUM_DEBT': 'sum',
    'AMT_CREDIT_SUM_OVERDUE': 'sum',
    'DAYS_CREDIT': 'mean',
    'DAYS_CREDIT_ENDDATE': 'mean'
}).reset_index()
```

```python
bureau_agg.columns = [
    'SK_ID_CURR',
    'BUREAU_ACTIVE_LOAN_RATIO',
    'BUREAU_TOTAL_CREDIT',
    'BUREAU_TOTAL_DEBT',
    'BUREAU_TOTAL_OVERDUE',
    'BUREAU_DAYS_CREDIT_MEAN',
    'BUREAU_DAYS_CREDIT_END_MEAN'
]
```

```python
## Merge bureau aggregated data with main dataframe
```

```python
df = df.merge(bureau_agg, on='SK_ID_CURR', how='left')
```

```python
## Fill missing bureau values with 0
```

```python
bureau_cols = [
    'BUREAU_ACTIVE_LOAN_RATIO',
    'BUREAU_TOTAL_CREDIT',
    'BUREAU_TOTAL_DEBT',
    'BUREAU_TOTAL_OVERDUE',
    'BUREAU_DAYS_CREDIT_MEAN',
    'BUREAU_DAYS_CREDIT_END_MEAN'
]

df[bureau_cols] = df[bureau_cols].fillna(0)
```

```python
df[bureau_cols].describe()
```

```python
prev_app.isnull().sum()
```

```python
## Merge previous application features
```

```python
df = df.merge(prev_agg, on='SK_ID_CURR', how='left')
```

```python
prev_cols = [
    'PREV_APPROVAL_RATE',
    'PREV_MEAN_APPLICATION_AMT',
    'PREV_MEAN_CREDIT_AMT',
    'PREV_DAYS_DECISION_MEAN',
    'PREV_APPLICATION_COUNT'
]

df[prev_cols] = df[prev_cols].fillna(0)
```

```python
## TARGET VARIABLE
```

```python
y = df['TARGET']
X = df.drop(columns=['TARGET'])
```

```python
## Identify categorical columns
```

```python
categorical_cols = X.select_dtypes(include=['object']).columns
numerical_cols = X.select_dtypes(exclude=['object']).columns

print("Categorical columns:", len(categorical_cols))
print("Numerical columns:", len(numerical_cols))
```

```python
## One-hot encode categorical variables
```

```python
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
```

```python
## Train–Test Split
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
## Logistic Regression (Baseline)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train)
```

```python
## Predictions & Evaluation
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred_log = log_model.predict(X_test)
y_prob_log = log_model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred_log))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_log))
```

```python
## Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
```

```python
## Evaluation
```

```python
y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
```

```python
## XGBoost
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    eval_metric='logloss',
    random_state=42
)

xgb_model.fit(X_train, y_train)
```

```python
## Evaluation
```

```python
y_pred_xgb = xgb_model.predict(X_test)
y_prob_xgb = xgb_model.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_xgb))
```

```python
## Model Comparison Table
```

```python
model_comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'XGBoost'],
    'ROC_AUC': [
        roc_auc_score(y_test, y_prob_log),
        roc_auc_score(y_test, y_prob_rf),
        roc_auc_score(y_test, y_prob_xgb)
    ]
})

model_comparison
```
